```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
datal5=data.tail()
```

```python
sns.set(rc={'figure.figsize':(3,3)})
sns.lineplot(x="Date",y="Confirmed",data=datal5)
plt.show()
```

```python
x = data['Confirmed'].sum()
y = data['Cured'].sum()
z= data['Deaths'].sum()
active= x-y
print('Total Confirmed cases =',x)
print('Total Cured cases =',y)
print('Total Active cases =',active)
print('Total Death cases =',z)
barp = sns.barplot(x=['Confirmed','Cured','Deaths','active'],y=[x,y,z,active])
barp.set_yticklabels(labels=(barp.get_yticks()*1).astype(int))
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Deaths'])
plt.title('Deaths Cases')
plt.show()
```

```python
corr = data.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
